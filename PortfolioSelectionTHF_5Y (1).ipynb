```python
#Load python libraries
import numpy as np
import pandas as pd
from pyomo.environ import *
from pyomo.opt import SolverFactory
from openpyxl import load_workbook

#Load the data source
import yfinance as yf
```

```python
#Set time period and interval 
per='5Y'
inter='1mo'
```

```python
#Fetch data for companies of interest from yahoo finance 
#Information Technology 
tech=pd.DataFrame()
tc_list=["INFY.BO","HCLTECH.BO","TCS.BO","TECHM.BO","WIPRO.BO","RELIANCE.BO","LTI.BO","MPHASIS.BO","MINDTREE.BO","CIGNITI.BO"]
for i in range(len(tc_list)):
  tc_comp= yf.download(tickers=tc_list[i], period=per, interval=inter)
  tc_comp=tc_comp.dropna()
  tc_comp[tc_list[i]]=""
  for j in range(len(tc_comp)):
    tc_comp[tc_list[i]][j]=(tc_comp['Adj Close'][j]/tc_comp['Adj Close'][j-1])-1
  tech=pd.concat([tech,tc_comp[tc_list[i]]],axis=1)

tech = tech.iloc[1:60 , :]
tech.head()
```

```python
#Healthcare
health_care=pd.DataFrame()
hc_list=["APOLLOHOSP.BO","FORTIS.BO","LALPATHLAB.BO","TRABI.BO","NH.BO","CADILAHC.BO","WOCKPHARMA.BO","MEDINOV.BO","TEJNAKSH.BO","SUNPHARMA.BO","AUROPHARMA.BO",
         "CIPLA.BO","APLLTD.BO","DRREDDY.BO","IPCALAB.BO","STAR.BO","LAURUSLABS.BO","GRANULES.BO","NATCOPHARM.BO","TORNTPHARM.BO"]
for i in range(len(hc_list)):
  hc_comp= yf.download(tickers=hc_list[i], period=per, interval=inter)
  hc_comp=hc_comp.dropna()
  hc_comp[hc_list[i]]=""
  for j in range(len(hc_comp)):
    hc_comp[hc_list[i]][j]=(hc_comp['Adj Close'][j]/hc_comp['Adj Close'][j-1])-1
  health_care=pd.concat([health_care,hc_comp[hc_list[i]]],axis=1)

health_care = health_care.iloc[1:60 , :]
health_care.head()
```

```python
#Financials
fin=pd.DataFrame()
fin_list=["AXISBANK.BO","HDFCBANK.BO","ICICIBANK.BO","BAJAJFINSV.BO","KOTAKBANK.BO","JMFINANCIL.BO","SBIN.BO","YESBANK.BO","IIFL.NS","PML.BO"]
for i in range(len(fin_list)):
  comp= yf.download(tickers=fin_list[i], period=per, interval=inter)
  comp=comp.dropna()
  comp[fin_list[i]]=""
  for j in range(len(comp)):
    comp[fin_list[i]][j]=(comp['Adj Close'][j]/comp['Adj Close'][j-1])-1
  fin=pd.concat([fin,comp[fin_list[i]]],axis=1)

fin = fin.iloc[1:60 , :]
fin.head()
```

```python
#Auto 
au=pd.DataFrame()
au_list=["ASHOKLEY.BO","TATAMOTORS.BO","EICHERMOT.BO","HEROMOTOCO.NS","M&M.BO","AMARAJABAT.BO","EXIDEIND.BO"]
for i in range(len(au_list)):
  au_comp= yf.download(tickers=au_list[i], period=per, interval=inter)
  au_comp=au_comp.dropna()
  au_comp[au_list[i]]=""
  for j in range(len(au_comp)):
    au_comp[au_list[i]][j]=(au_comp['Adj Close'][j]/au_comp['Adj Close'][j-1])-1
  au=pd.concat([au,au_comp[au_list[i]]],axis=1)

au = au.iloc[1:60 , :]
au.head()
```

```python
au=pd.DataFrame()
au_list=["ASHOKLEY.BO","TATAMOTORS.BO","EICHERMOT.BO","HEROMOTOCO.NS","M&M.BO","AMARAJABAT.BO","EXIDEIND.BO"]
for i in range(len(au_list)):
  au_comp= yf.download(tickers=au_list[i], period=per, interval=inter)
  au_comp=au_comp.dropna()
  au_comp[au_list[i]]=""
  for j in range(len(au_comp)):
    au_comp[au_list[i]][j]=(au_comp['Adj Close'][j]/au_comp['Adj Close'][j-1])-1
  au=pd.concat([au,au_comp[au_list[i]]],axis=1)

au = au.iloc[1:60 , :]
au.head()
```

```python
#Write a function optimizes portofolio selection
def portfolio(threshold,sec1, sec2, sec3):
  df=pd.DataFrame()
  sectors=[sec1,sec2,sec3]
  
  for i in range(len(sectors)):
    if sectors[i]== "Healthcare":
      df=pd.concat([df,health_care],axis=1)
    if sectors[i]== "Tech":
      df=pd.concat([df,tech],axis=1)
    if sectors[i]== "Financials":
      df=pd.concat([df,fin],axis=1)
    if sectors[i]== "Auto":
      df=pd.concat([df,au],axis=1)

  data=df.values.tolist()  
  num_companies=len(df.columns) #indexed by i
  num_months=len(data) #indexed by j
  
  #Declare a model
  model=ConcreteModel()

  #Declare decision variables 
  model.x=Var(range(num_companies), domain=NonNegativeReals) # % contribution to portfolio
  model.y=Var(range(num_months),range(2),domain=NonNegativeReals) #pos-neg

  ret_month=[]
  for j in range(num_months):
    ret_month.append(sum(model.x[i]*data[j][i] for i in range(num_companies)))

  avg_return=sum(ret_month)/len(ret_month)
  dev=[]
  for j in range(num_months):
    dev.append(avg_return-ret_month[j]) 

  #Constraints
  #pos-neg
  model.constraint1=ConstraintList()
  for j in range(num_months):
    model.constraint1.add(expr= model.y[j,0]- model.y[j,1] == dev[j])

  #total percentage = 100%
  model.constraint2=Constraint(expr=sum(model.x[i] for i in range(num_companies))== 100)

  #threshold for MAD
  model.constraint3=Constraint(expr=sum(model.y[j,0]+model.y[j,1] for j in range(num_months))/num_months<= threshold)
  
  #more than 30% cannot be invested in one company 
  model.constraint4=ConstraintList()
  for i in range(num_companies):
    model.constraint4.add(expr=model.x[i]<=30)
  
  #Set Objective: Maximize returns 
  model.Objective=Objective(expr=sum(ret_month)/num_months, sense=maximize)
  
  #Solve
  opt=SolverFactory('glpk')
 #opt.options["mipgap"]=0.001 #tolerance
  results=opt.solve(model, tee=True)

  #Calculation for Sharpe ratio
  var=sum((model.y[j,0]()+model.y[j,1]())**2 for j in range(num_months))/num_months
  mad=sum(model.y[j,0]()+model.y[j,1]() for j in range(num_months))/num_months
  sharpe=model.Objective()/var**0.5

  print("Average Return is", round(model.Objective(),3), "%")
  print("Average MAD is", round(mad,3),"%")
  print("Average Variance is", round(var,3),"%")
  print("\nThe capital should be distributed as follows:")
  col_names = df.columns.values
  for i in range(num_companies):
    if model.x[i]()!=0: 
      print(round(model.x[i](),3),"% in",col_names[i])

  print("\n This investement has a Sharpe ratio of", round(sharpe,3))
```
